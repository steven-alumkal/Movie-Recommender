```python
import pandas as pd

def recommend_movies(data):
    # Step 1: Get user preferences
    print("Welcome to the Movie Recommender!")
    preferred_genre = input("What genre do you prefer? (e.g., Action, Comedy, Drama, etc.): ")
    preferred_score = input("What is your minimum score preference? (e.g., Low, Medium, High): ")

    # Step 2: Map user input to numeric values for filtering
    score_mapping = {'Low': (0, 40), 'Medium': (40, 70), 'High': (70, 100)}
    if preferred_score not in score_mapping:
        print("Invalid score preference. Please choose from Low, Medium, or High.")
        return

    # Convert genres to strings and fill NaNs with an empty string
    data['genres'] = data['genres'].astype(str).fillna('')

    # Step 3: Filter the DataFrame based on user input
    score_range = score_mapping[preferred_score]
    filtered_data = data[
        (data['score'] >= score_range[0]) &
        (data['score'] < score_range[1]) &
        (data['genres'].str.contains(preferred_genre, case=False))
    ]

    # Debugging output
    print(f"Score Range: {score_range}")
    print(f"Preferred Genre: {preferred_genre}")
    print(f"Filtered Data Shape: {filtered_data.shape}")
    
    # Step 4: Recommend top 5 movies based on the filtered data
    if filtered_data.empty:
        print("No movies found matching your criteria.")
    else:
        top_movies = filtered_data.nlargest(5, 'score')
        print("Top 5 recommended movies:")
        for index, row in top_movies.iterrows():
            print(f"- {row['original_title']} (Score: {row['score']})")

# Example usage:
recommend_movies(data)
```
